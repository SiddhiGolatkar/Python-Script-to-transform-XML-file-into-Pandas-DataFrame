```python
def parse_xml(root):
    attr = root.attrib
    for root_element in myroot.findall("document"):
        xml_data = attr.copy()
        xml_data['docid'] = root_element.find('docid').text
        xml_data['source'] = root_element.find('source').text
        xml_data['url'] = root_element.find('url').text
        xml_data['title'] = root_element.find('title').text
        xml_data['date'] = root_element.find('date').text
        yield xml_data
```

```python
parse_xml(myroot)
```

```python
data = parse_xml(myroot)
```
